```python
#importing libraries
import pandas as pd
import numpy as np

```

```python
#loading dataset
df = pd.read_excel('Business_Funding_Data.xlsx')
```

```python
#number of rows
len(df)
```

```python
# Percentage of missing values
missing_percentage = df.isnull().sum()/len(df)*100
missing_percentage
```

```python
#dropping the 'effective date' column as it is irrelevant
df.drop(columns=('Effective date'), inplace=True)
df
```

```python
#dropping the 'Financing Type' column as it is irrelevant
df.drop(columns=('Financing Type'), inplace=True)
df
```

```python
#dropping the 'Found At' column as it is irrelevant
df.drop(columns=('Found At'), inplace=True)
df
```

```python
#Filling null entries in 'Financing Type Normalized' column with 'Not Provided'
df['Financing Type Normalized'] = df['Financing Type Normalized'].fillna('Not Provided')
```

```python
df
```

```python
#Filling null entries in 'Investors' column with 'Not Provided'
df['Investors'] = df['Investors'].fillna('Not Provided')
```

```python
df
```

```python
#Filling null entries in 'Investors Count' column with 'Zero'
df['Investors Count'] = df['Investors Count'].fillna(0)
df
```

```python
#Filling null entries in 'Categories' column with 'Not Provided'
df['Categories'] = df['Categories'].replace('[]','["Not Provided"]')
```

```python
df
```

```python
#Checking duplicate rows
duplicates = df.duplicated().sum()
print(duplicates)
```

```python
#Confirming if there are no more null values
df.info()
```

```python
#All null values have been well handled, data is ready for further analysis
# Export the cleaned dataset to a new CSV file
df.to_csv('cleaned_Business_Funding_Data.csv', index=False)
print("Cleaned dataset exported to 'cleaned_Business_Funding_Data.csv'")
```
